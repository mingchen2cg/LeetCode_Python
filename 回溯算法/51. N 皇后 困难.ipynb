```python
class Solution(object):
    def solveNQueens(self, n):
        """
        :type n: int
        :rtype: List[List[str]]
        """
        self.result = []
        self.chessboard = [['.'] * n for _ in range(n)]
        self.backtrack(n,0)
        
        return self.result
    
    
    def backtrack(self,n,row):
        if row == n:
            self.result.append(["".join(r) for r in self.chessboard])  
            return
        
        for col in range(n):
            if self.isValid(row,col,n):
                self.chessboard[row][col] = 'Q'
                self.backtrack(n,row+1)
                self.chessboard[row][col] = '.'
        
    def isValid(self,row,col,n):
        # 检查列
        for i in range(row):
            if self.chessboard[i][col] == 'Q':
                return False
        
        # 检查45度
        i = row - 1
        j = col - 1
        while i >= 0 and j >= 0:  # 只要不越界
            if self.chessboard[i][j] == 'Q':  # 检查是否有皇后
                return False
            i -= 1  # 往左上方移动
            j -= 1
        
        # 检查45度
        i = row - 1
        j = col + 1 
        while i >=0 and j <n:
            if self.chessboard[i][j] == 'Q':
                return False
            i -= 1 
            j += 1 
            
        return True
```

```python
class Solution(object):
    def solveNQueens(self, n):
        """
        :type n: int
        :rtype: List[List[str]]
        """
        self.n = n
        self.results = []
        # 初始化棋盘，每个位置用 '.' 表示
        board = [['.'] * n for _ in range(n)]
        self.dfs(0, board)
        return self.results

    def dfs(self, row, board):
        # 如果所有行都放置了皇后，则找到一个解
        if row == self.n:
            self.results.append(["".join(r) for r in board])
            return

        for col in range(self.n):
            if self.is_valid(board, row, col):
                # 尝试在 (row, col) 放置皇后
                board[row][col] = 'Q'
                self.dfs(row + 1, board)
                # 回溯，撤销放置
                board[row][col] = '.'

    def is_valid(self, board, row, col):
        # 检查同一列是否已有皇后
        for i in range(row):
            if board[i][col] == 'Q':
                return False

        # 检查左上对角线上是否已有皇后
        i, j = row - 1, col - 1
        while i >= 0 and j >= 0:
            if board[i][j] == 'Q':
                return False
            i -= 1
            j -= 1

        # 检查右上对角线上是否已有皇后
        i, j = row - 1, col + 1
        while i >= 0 and j < self.n:
            if board[i][j] == 'Q':
                return False
            i -= 1
            j += 1

        return True
```
