```python
class Solution(object):
    def restoreIpAddresses(self, s):
        """
        :type s: str
        :rtype: List[str]
        """
        n = len(s)
        result = []

        def is_valid_ip_segment(s):
            """
            判断字符串 s 是否可以表示一个 0 到 255 之间的整数，且不含有前导 0。
            :param s: 输入的字符串
            :return: 如果合法返回 True，否则返回 False
            """
            n = len(s)
            # 规则 1：长度检查
            if n == 0 or n > 3:
                return False
            # 规则 2：前导 0 检查
            if n > 1 and s[0] == '0':
                return False
            # 规则 3：数值范围检查
            num = int(s)
            return 0 <= num <= 255
        
        def backtrack(start, path):
            # 如果路径长度已经达到4，且已经遍历完整个字符串，则加入结果
            if len(path) == 4:
                if start == n:
                    result.append(".".join(map(str, path)))
                return  # 无论是否满足条件，都返回，因为路径长度已经达到4

            # 剪枝：如果剩余字符过多或过少，直接返回
            remaining_length = n - start
            min_required = 4 - len(path)  # 每段至少1位
            max_required = 3 * (4 - len(path))  # 每段最多3位
            if remaining_length < min_required or remaining_length > max_required:
                return

            # 尝试分割字符串
            for end in range(start + 1, min(start + 4, n + 1)):  # 每段最多3位
                subipstr = s[start:end]
                if is_valid_ip_segment(subipstr):
                    path.append(subipstr)
                    backtrack(end, path)
                    path.pop()
                    
        backtrack(0, [])
        return result
```
