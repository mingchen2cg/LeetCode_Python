```python
import sys
input = sys.stdin.read

def main():
    ans = sys.maxsize
    data = input().split()
    index = 0
    # n, m 的输入处理  
    n = int(data[index])
    index += 1 
    m = int(data[index])
    index += 1

    value_matrix = [[0 for _ in range(m)] for _  in range(n)]

    # 地块价值的输入处理  
    for i in range(n) :
        for j in range(m):
            value_matrix[i][j] = int(data[index])
            index += 1
    
    hang_sum = [0] * n
    lie_sum = [0] * m
    for i in range(n):
        hang_sum[i] = sum(value_matrix[i])
    for i in range(m):
        lie = 0
        for j in range(n):
            lie += value_matrix[j][i]
        lie_sum[i] = lie
    
    
    hang_preSum = [0] * n
    lie_preSum = [0] * m
    tmp = 0 
    for i in range(n):
        
        tmp += hang_sum[i]
        hang_preSum[i] += tmp
    tmp = 0 
    for i in range(m):
        tmp += lie_sum[i]
        lie_preSum[i] += tmp
    
    

    # 行搜索
    for i in range(n):
        chajia = abs(hang_preSum[n - 1] - (hang_preSum[i] * 2))
        ans = min(chajia, ans)
    # 列搜索
    for i in range(m):
        chajia = abs(lie_preSum[m - 1] - (lie_preSum[i] * 2))
        ans = min(chajia, ans)

    print(ans)


if __name__ == "__main__":
    main()
```
