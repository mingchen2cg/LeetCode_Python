```python
from collections import deque

class Solution(object):
    def reverseOddLevels(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: Optional[TreeNode]
        """
        if not root:
            return root
        
        queue = deque()
        queue.append(root)
        depth = 0
        
        while queue:
            level_size = len(queue)
            
            # 如果是奇数层，交换节点值
            if depth % 2 != 0:
                # 使用双指针交换节点值
                left = 0
                right = level_size - 1
                while left < right:
                    # 交换节点值
                    queue[left].val, queue[right].val = queue[right].val, queue[left].val
                    left += 1
                    right -= 1
            
            # 将下一层的节点加入队列
            for _ in range(level_size):
                current_node = queue.popleft()
                if current_node.left:
                    queue.append(current_node.left)
                if current_node.right:
                    queue.append(current_node.right)
            
            depth += 1
        
        return root
```

```python
class Solution(object):
    def reverseOddLevels(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: Optional[TreeNode]
        """

        self.reverse_tree(root.left, root.right, True)
        return root

    def reverse_tree(self, root1, root2, is_reverse):
        if root1 is None:       # 由于是完全二叉树，写一个判断左子树或者右子树是否为空的语句即可
            return

        if root2 is None:
            return

        if is_reverse:                      # 奇数层镜像交换节点内的数值（相当于反转）
            root1.val, root2.val = root2.val, root1.val

        is_reverse = not is_reverse

        self.reverse_tree(root1.left, root2.right, is_reverse)      # 镜像递归左右子树并向下传递奇偶层的信息
        self.reverse_tree(root1.right, root2.left, is_reverse)      # 镜像递归左右子树并向下传递奇偶层的信息
```
