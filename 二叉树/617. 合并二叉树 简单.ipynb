### 递归思路：
> 1. **基本情况**：
>    - 如果 `t1` 为空，返回 `t2`。
>    - 如果 `t2` 为空，返回 `t1`。
>    
> 2. **递归步骤**：
>    - 创建一个新节点，其值为 `t1` 和 `t2` 当前节点值的和。
>    - 递归地合并 `t1` 和 `t2` 的左子树，并将结果作为新节点的左子树。
>    - 递归地合并 `t1` 和 `t2` 的右子树，并将结果作为新节点的右子树。
>    
> 3. **返回结果**：
>    - 返回新创建的节点。

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution(object):
    def mergeTrees(self, root1, root2):
        """
        :type root1: Optional[TreeNode]
        :type root2: Optional[TreeNode]
        :rtype: Optional[TreeNode]
        """
        if not root1:
            return root2
        
        if not root2:
            return root1
        
        root = TreeNode(root1.val + root2.val)
        
        root.left = self.mergeTrees(root1.left, root2.left)
        root.right = self.mergeTrees(root1.right, root2.right)
        
        return root
```

> ### 迭代方法思路
> 
> 1. **基本情况**：
>    - 如果 `t1` 为空，直接返回 `t2`。
>    - 如果 `t2` 为空，直接返回 `t1`。
> 
> 2. **初始化**：
>    - 使用一个队列来存储需要合并的节点对（`t1` 和 `t2` 的对应节点）。
> 
> 3. **迭代过程**：
>    - 从队列中取出一个节点对。
>    - 将 `t2` 的节点值加到 `t1` 的节点值上。
>    - 如果 `t1` 和 `t2` 都有左子节点，将它们的左子节点对加入队列。
>    - 如果 `t1` 没有左子节点，直接将 `t2` 的左子节点赋值给 `t1`。
>    - 如果 `t1` 和 `t2` 都有右子节点，将它们的右子节点对加入队列。
>    - 如果 `t1` 没有右子节点，直接将 `t2` 的右子节点赋值给 `t1`。
> 
> 4. **返回结果**：
>    - 返回合并后的 `t1` 树。

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def mergeTrees(self, root1, root2):
        """
        :type root1: Optional[TreeNode]
        :type root2: Optional[TreeNode]
        :rtype: Optional[TreeNode]
        """
        if not root1:
            return root2
        
        if not root2:
            return root1
        
        from collections import deque
        
        queue = deque()
        queue.append((root1, root2))
        
        while queue:
            node1, node2 = queue.popleft()
            node1.val += node2.val
            
            if node1.left and node2.left:
                queue.append((node1.left, node2.left))
            
            elif not node1.left:
                node1.left = node2.left
                
            if node1.right and node2.right:
                queue.append((node1.right, node2.right))
            
            elif not node1.right:
                node1.right = node2.right

        return root1
        
```

```python

# 最快的方法

class Solution(object):
    def mergeTrees(self, root1, root2):
        """
        :type root1: Optional[TreeNode]
        :type root2: Optional[TreeNode]
        :rtype: Optional[TreeNode]
        """
        # 如果 root1 为空，返回 root2
        if not root1:
            return root2
        # 如果 root2 为空，返回 root1
        if not root2:
            return root1
        
        # 合并当前节点的值
        root1.val += root2.val
        
        # 递归合并左子树
        root1.left = self.mergeTrees(root1.left, root2.left)
        # 递归合并右子树
        root1.right = self.mergeTrees(root1.right, root2.right)
        
        # 返回合并后的树
        return root1
```
