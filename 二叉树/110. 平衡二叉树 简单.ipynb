```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isBalanced(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        def checkHeight(node):
            # 如果节点为空，返回高度0
            if not node:
                return 0
            
            # 递归计算左子树的高度
            left_height = checkHeight(node.left)
            # 如果左子树不平衡，返回-1
            if left_height == -1:
                return -1
            
            # 递归计算右子树的高度
            right_height = checkHeight(node.right)
            # 如果右子树不平衡，返回-1
            if right_height == -1:
                return -1
            
            # 如果左右子树高度差超过1，返回-1表示不平衡
            if abs(left_height - right_height) > 1:
                return -1
            
            # 返回当前节点的高度
            return max(left_height, right_height) + 1
        
        # 如果checkHeight返回-1，说明树不平衡，否则平衡
        return checkHeight(root) != -1
        
        
```

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def isBalanced(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """
        return self.checkBalance(root) != -1
    
    def checkBalance(self,node):
        if not node:
            return 0
        
        left_height = self.checkBalance(node.left)
        if left_height == -1:
            return -1
        
        right_height = self.checkBalance(node.right)
        if right_height == -1:
            return -1 
        
        if abs(right_height - left_height) > 1 :
            return -1 
        
        return max(left_height, right_height) + 1
        
```

```python
# 定义二叉树节点类
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
# 辅助函数：根据列表构建二叉树
def build_tree(nodes, index=0):
    if index >= len(nodes) or nodes[index] is None:
        return None
    root = TreeNode(nodes[index])
    root.left = build_tree(nodes, 2 * index + 1)
    root.right = build_tree(nodes, 2 * index + 2)
    return root
```
