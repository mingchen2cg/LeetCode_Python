```python
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        
        if len(s) < len(t):
            return ""
         # 统计一个字符串中每个字符的频次
        def count_characters(str):
            char_count = {}
            for char in str:
                char_count[char] = char_count.get(char, 0) + 1 
            return char_count
        
        t_count = count_characters(t)  # t 中每个字符的需求数量
        window_count = {}               # 当前窗口中的字符计数

        required = len(t_count)  # t 中不同字符的种类数
        formed = 0  # 当前窗口中满足 t 需求的字符种类数
        i = 0       # 左指针
        min_len = float("inf")  # 最小子串长度
        min_window = ""  # 记录最小覆盖子串

        for j in range(len(s)):
            # 将右边字符加入窗口
            char = s[j]
            window_count[char] = window_count.get(char,0) + 1 

            # 检查该字符是否满足 t 中的需求
            if char in t_count and window_count[char] == t_count[char]:
                formed += 1

            # 缩小窗口，直到不满足条件
            while i <= j and formed == required:
                char = s[i]

                # 更新最小子串
                if j - i + 1 < min_len:
                    min_len = j - i + 1
                    min_window = s[i:j+1]

                # 移出窗口左边的字符
                window_count[char] -= 1
                if char in t_count and window_count[char] < t_count[char]:
                    formed -= 1
                
                i += 1 

        return min_window





        
```

```python
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
                
        if len(s) < len(t):
            return ""
        # 统计一个字符串中每个字符的频次
        def count_characters(str):
            char_count = {}
            for char in str:
                char_count[char] = char_count.get(char, 0) + 1 
            return char_count
        # 初始声明变量
        i = 0 # 左指针 
        
        formed = 0 # 滑动窗口中已经满足要求的字符数量
        t_dict = count_characters(t)  # t中每个字符需要的数量 
        window_dict = {}
        min_len = float("inf")  # 最小子串长度
        min_window = ""  # 滑动窗口中的字符 

        for j in range(len(s)):
            # 右指针右移
            char = s[j]
            window_dict[char] = window_dict.get(char, 0) + 1 

            # 检查该字符是否满足条件
            if char in t_dict.keys() and window_dict[char] == t_dict[char]:
                formed += 1 
            
            while formed == len(t_dict) and i <= j:
                char = s[i]
                 # 更新最小子串
                if j - i + 1 < min_len:
                    min_len = j - i + 1
                    min_window = s[i:j+1]

                # 移出窗口左边的字符
                window_dict[char] -= 1
                if char in t_dict and window_dict[char] < t_dict[char]:
                    formed -= 1
                
                i += 1
        
        return min_window


```
