```python
class Solution(object):
    def fourSumCount(self, nums1, nums2, nums3, nums4):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :type nums3: List[int]
        :type nums4: List[int]
        :rtype: int
        """

        cnt = 0

        ab = {}
        for i in nums1:
            for j in nums2:
                ab[i+j] = ab.get(i+j,0) + 1
        
        for i in nums3:
            for j in nums4:
                key = - (i + j)
                cnt += ab.get(key, 0)  
                     
        return cnt
```
